```python
import pandas as pd
import numpy as np
```

```python
cell_phones = pd.read_csv("https://raw.githubusercontent.com/zalihat/cell_phone_dataset/master/cell_phones.csv")
population = pd.read_csv("https://raw.githubusercontent.com/zalihat/cell_phone_dataset/master/population1.csv")
```

```python
country_codes = pd.read_html("https://github.com/open-numbers/ddf--gapminder--systema_globalis/blob/master/ddf--entities--geo--country.csv")
country_codes = country_codes[0]
```

```python
phones_population = pd.merge(cell_phones,population, how = 'inner')
phones_population
```

```python
country_codes = country_codes[['country','name']].rename(columns={'country':'geo', 'name':'Country'})
phones_population_codes = pd.merge(phones_population,country_codes)
phones_population_codes
```

```python
phones_population_codes['cell_phones_per_person'] = phones_population_codes['cell_phones_total']/phones_population_codes['population']
phones_population_codes
```

```python
phones_population_codes.loc[phones_population_codes['Country']== 'United States']
```

```python
phones_population_codes['geo'] = phones_population_codes['geo'].str.upper()
phones_population_codes
```
